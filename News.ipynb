```python
import requests
from lxml import html
from lxml import etree
import pandas as pd
from datetime import datetime
```

# Парсим lenta.ru:

```python
current = datetime.today().strftime('%Y-%m-%d')

url = 'https://lenta.ru'
header = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}
response = requests.get(url, headers=header)
response.encoding = 'utf-8'

myparser = etree.HTMLParser(encoding="utf-8")
dom = etree.HTML(response.content, parser=myparser)

list_1 = dom.xpath('/html/body/div[@class = "layout js-layout"]/div[@class="layout__container"]/main/div/section[1]/div[1]/div[@class="topnews"]/*[@class="topnews__column"]/a/@href')
list_2 = dom.xpath('/html/body/div[@class = "layout js-layout"]/div[@class="layout__container"]/main/div/section[1]/div[1]/div[@class="topnews"]/*[@class="topnews__column"]/a/div/span/text()')
list_3 = dom.xpath('/html/body/div[@class = "layout js-layout"]/div[@class="layout__container"]/main/div/section[1]/div[1]/div[@class="topnews"]/*[@class="topnews__column"]/a/div/div/time/text()')

main_href = dom.xpath('/html/body/div[@class = "layout js-layout"]/div[@class="layout__container"]/main/div/section[1]/div[1]/div[1]/div[1]/div[1]/a/@href')
main_name = dom.xpath('/html/body/div[@class = "layout js-layout"]/div[@class="layout__container"]/main/div/section[1]/div[1]/div[1]/div[1]/div[1]/a/div[2]/h3/text()')
main_time = dom.xpath('/html/body/div[@class = "layout js-layout"]/div[@class="layout__container"]/main/div/section[1]/div[1]/div[1]/div[1]/div[1]/a/div[3]/time/text()')

list_1.insert(0, main_href[0])
list_2.insert(0, main_name[0])
list_3.insert(0, main_time[0])

list_4 = ['other' for el in list_1]
for i, el in enumerate(list_1):
    if el.startswith('/news'):
        list_1[i] = 'https://lenta.ru' + el
        list_4[i] = 'Lenta.ru'
        
for i, el in enumerate(list_3):
    list_3[i] = current + ':' + str(el)
    
df = pd.DataFrame({'Name': list_2, 'href': list_1, 'time': list_3, 'source': list_4})
df
```
